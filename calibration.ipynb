```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('20191206173609-1324367673.json') as json_file:
    data = json.load(json_file)
barlovic = pd.io.json.json_normalize(data)
```

```python
with open('20191202134100-1575312060582.json') as json_file:
    data = json.load(json_file)
ai = pd.io.json.json_normalize(data)
```

```python
g = sns.lmplot("Detector.totalDensity", "Detector.totalFlow", data=barlovic[barlovic['Detector.totalDensity']<80], hue="Detector.idNum", markers=[".", "+", "x", "1"],
          fit_reg=False, legend=False)
#g.set(xlim=(0,50))
plt.legend(title="Detector", labels=['-1 km','-3 km', "-5 km", "-7 km"])
plt.xlabel("Density (1 / km)")
plt.ylabel("Flow (1 / h)")
plt.show(g)
```

```python
g = sns.lmplot("Detector.totalDensity", "Detector.totalFlow", data=ai[ai['Detector.totalDensity']<80], hue="Detector.idNum", markers=[".", "+", "x", "1"],
          fit_reg=False, legend=False)
#g.set(xlim=(0,50))
plt.legend(title="Detector", labels=['-1 km','-3 km', "-5 km", "-7 km"])
plt.xlabel("Density (1 / km)")
plt.ylabel("Flow (1 / h)")
plt.show(g)
```
