```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'pdf'
```

```python
with open('hdv_two_lane.json') as json_file:
    data = json.load(json_file)
barlovic = pd.io.json.json_normalize(data)
```

```python
g = sns.lmplot("Detector.totalDensity", "Detector.totalFlow", data=barlovic, hue="Detector.idNum", markers=[".", "+", "x", "1"],
          fit_reg=False, legend=False)
#g.set(xlim=(0,50))
plt.legend(title="Detector", labels=['-1 km','-3 km', "-5 km", "-7 km"])
plt.xlabel("Density (veh / km)")
plt.ylabel("Flow (veh / h)")
plt.show(g)
```
